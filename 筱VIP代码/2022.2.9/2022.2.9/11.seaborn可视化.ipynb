```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = pd.read_csv("./iris.csv")
iris.head()
```

```python
plt.figure(1,figsize=(12,6))
plt.subplot(1,2,1)
sns.stripplot(x='Species',y='Sepal.Length',data=iris)
plt.title('Striplot of sepal length of Iris species')

with sns.axes_style("whitegrid"):
    plt.subplot(1,2,2)
    plt.title('Striplot of sepal length of Iris species')
    sns.stripplot(x='Species',y='Sepal.Length',data=iris)
```

```python
plt.figure(1,figsize=(12,6))
plt.subplot(1,2,1)
sns.swarmplot(x='Species',y='Petal.Length',data=iris) 

with sns.axes_style("ticks"): # ticks
    plt.subplot(1,2,2)
    sns.swarmplot(x='Species',y='Petal.Width',data=iris)

plt.show()
```

```python
var = ['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']
axes_style = ['ticks','white','whitegrid', 'dark']

fig = plt.figure(1,figsize=(12,12))

for i in range(4):
    with sns.axes_style(axes_style[i]): # darkgrid
        plt.subplot(2,2,i+1)
        sns.boxplot(x='Species',y=var[i],data=iris)

plt.show()
```

```python
#Violinplot
context= ['notebook','paper','talk','poster']
axes_style = ['ticks','white','whitegrid', 'dark']

plt.figure(1,figsize=(12,12))
for i in range(4):
    with sns.axes_style(axes_style[i]):#axes_style
        sns.set_context(context[i]) # context style,notebook,paper,talk,poster
        plt.subplot(2,2,i+1)
        plt.title(str(var[i])+ ' in Iris species')
        sns.violinplot(x='Species',y=var[i],data=iris)

plt.show()
```

```python
#Violinplotkernel density estimatequantitative
context= ['notebook','paper','talk','poster']
axes_style = ['ticks','white','whitegrid', 'dark']

plt.figure(1,figsize=(12,12))
for i in range(4):
    with sns.axes_style(axes_style[i]):#axes_style
        sns.set_context(context[i])#context
        plt.subplot(2,2,i+1)
        plt.title(str(var[i])+ ' in Iris species')
        sns.swarmplot(x='Species', y=var[i], data=iris, color="w", alpha=.5) 
        sns.violinplot(x='Species', y=var[i], data=iris, inner=None) if i%2 ==0 \
        else sns.boxplot(x='Species', y=var[i], data=iris) # swarmplot+violinplot swarmplot + boxplot

plt.show()
```

```python
plt.figure(1,figsize=(12,12))
for i in range(4):
    with sns.axes_style(axes_style[i]):#axes_style
        sns.set_context(context[i]) # context style,notebook,paper,talk,poster
        plt.subplot(2,2,i+1)
        plt.title(str(var[i])+ ' in Iris species')
        sns.barplot(x='Species',y=var[i],data=iris)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(y="Species", data=iris) # y='species',countplot
plt.title('Iris species count')
plt.show()
```

```python
plt.figure(1,figsize=(12,12))
for i in range(4):
    with sns.axes_style(axes_style[i]):#axes_style
        sns.set_context(context[i]) # context style,notebook,paper,talk,poster
        plt.subplot(2,2,i+1)
        plt.title(str(var[i])+ ' in Iris species')
        sns.pointplot(x='Species',y=var[i],data=iris)
plt.show()
```
